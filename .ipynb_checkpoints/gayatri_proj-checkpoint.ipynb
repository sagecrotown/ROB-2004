# Code Skeleton for the simulation

Feel free to change the `run_time` variable to match your needs. The class `Simulator` contains all the code for the simulation. The code skeleton currently implement a simple PD controller in joint space that just keeps a desired current position.

Importantly you can control both the joints and the gripper of the robot:
* `simulator.send_joint_torque(joint_torques)` sends a vector of dimension 7 setting the torques of the joints
* `simulator.gripper_move(gripper_position)` sends a vector of dimension 2 setting the position of the fingers of the gripper. The fingers can have positions from 0 to 0.04cm (you can fully close the gripper by setting `simulator.gripper_move([0,0])` and fully open it by setting `simulator.gripper_move([0.04,0.04)`)

```python
def compute_trajectory(position_init, position_goal, t_init, t_goal, t):
    p_dif, t_dif = position_goal - position_init, t - t_init
    T = t_goal - t_init
    a3, a4, a5 = 10/T**3, -15/T**4, 6/T**5
    desired_position = position_init + ( a3*t_dif**3 + a4*t_dif**4 + a5*t_dif**5 ) * ( p_dif )
    desired_velocity = ( 3*a3*t_dif**2 + 4*a4*t_dif**3 + 5*a5*t_dif**4 ) * ( p_dif )
    return desired_position, desired_velocity
```

```python
# here we create a simulation and reset the state of the robot
simulator = Simulator()
simulator.reset_state([0.,0,0.0,-np.pi/2.,0.,np.pi/2.,0.0])

# we create a robot object so we can use its helper functions
my_robot = PandaRobot()

# duration of the simulation
run_time = 30.

# simulation time step
dt = 0.005
num_steps = int(run_time/dt)

time_to_goal = 7.5
target_num = 0
t = 0.
eps = 10e-4


# target positions
block1_pos = np.array([0.35,0.58,0.65])
block2_pos = np.array([0.15,0.67,0.65])
bowl_pos = np.array([-0.3,0.55,0.65])
targets = [block1_pos, bowl_pos, block2_pos, bowl_pos]

# open claw to start
gripper_position = [0.4, 0.4]

# we store information
ndofs = 7 # number of degrees of freedom (excluding the gripper)
time = np.zeros([num_steps]) # time stamps

measured_joint_positions = np.zeros([num_steps,ndofs]) # measured joint positions
measured_end_eff_positions = np.zeros([num_steps,ndofs]) # measured end-effector positions

measured_joint_velocities = np.zeros_like(measured_joint_positions) # measured joint velocities
measured_end_eff_velocities = np.zeros_like(measured_joint_positions) # measured end-effector velocities

desired_joint_positions = np.zeros_like(measured_joint_positions) # desired joint positions
desired_end_eff_positions = np.zeros_like(measured_joint_positions) # desired end-effector positions

desired_joint_velocities = np.zeros_like(measured_joint_positions) # desired joint velocities
desired_end_eff_velocities = np.zeros_like(measured_joint_positions) # desired end-effector velocities

desired_torques = np.zeros_like(measured_joint_positions) # torques applied

### this is the main control loop
for i in range(num_steps):
    # get the current time and save it
    time[i] = dt * i
    
    # we get the position and velocities of the joints
    q, dq = simulator.get_state()    
    end_eff_pose = my_robot.FK(q)
    jacobian = my_robot.get_jacobian(q,'O')
    end_eff_vel = jacobian[3:] @ dq
    
        
    # the PD gains
    # P = np.array([100., 100., 100., 100., 100., 100., 50.])
    P = np.array([100., 100., 100.])
    D = np.array([10.,10,10,10,10,10,1.])
    
    # determine trajectory
    
    # determine init_pos and goal_pos
    if time[i] == 0:
        init_pos = end_eff_pose[:3,3]
    elif time_to_goal - t <= 0:
        init_pos = targets[target_num]
        if (target_num%2 == 0): #open gripper
            gripper_position = [0.1, 0.1]
        else: #open
            gripper_position = [0.4, 0.4]
        target_num += 1
        t=0
    goal_pos = targets[target_num]
    
    # calculate trajectory    
    q_des, dq_des = compute_trajectory(init_pos, goal_pos, 0, time_to_goal, t)
    
    # p control
    
    v_0 = (dq_des + P*(q_des - end_eff_pose[:3,3]))
    
    # inverse jacobian
    invJacobian = pinv(jacobian[3:])
    dtheta_des = invJacobian @ v_0
    d_error = dtheta_des - dq
    joint_torques = (D*d_error) + my_robot.g(q)
    
    # save    
    measured_joint_positions[i,:] = q
    measured_joint_velocities[i,:] = dq
    #measured_end_eff_positions[i,:] = end_eff_pose[:3,3]
    #measured_end_eff_velocity[i,:] = end_eff_vel
    #desired_end_eff_positions[i,:] = q_des
    #desired_end_eff_velocities[i,:] = dq_des
    desired_torques[i,:] = joint_torques
    
    
#     singularity_orientation = (abs(q[2]) < 0.05) and (abs(q[4]) < 0.05) and (abs(q[6]) < 0.05)
#     if singularity_orientation:
#         invJacobian = pinv(jacobian[3:] + eps*np.eye(3))
#     else:         
    
    ## controller: TODO HERE IMPLEMENT YOUR CONTROLLER TO SOLVE THE TASK
    ## you will need to replace the naive PD controller implemented below

#     # desired joint positions and velocities
#     q_des = np.array([0.,0,0.0,-np.pi/2.,0.,np.pi/2,0.0])
#     dq_des = np.zeros([7])
    
#     # we save the desired positions/velocities for later plotting 
#     desired_positions[i,:] = q_des
#     desired_velocities[i,:] = dq_des
    
#     ##PD controller
    

#     error = q_des - q # the position error for all the joints
#     d_error = dq_des-dq # the velocity error for all the joints
    
    
#     # we compute the desired torques as a PD controller
#     joint_torques = np.diag(P) @ error + np.diag(D) @ d_error + my_robot.g(q)
    
    # we start by closing the gripper and then open it after 5 seconds
#     gripper_position = [0,0]
#     if time[i] > 5.0:
#         gripper_position = [0.04, 0.04]
    
    # we send them to the robot and do one simulation step
    simulator.send_joint_torque(joint_torques)
    simulator.gripper_move(gripper_position)
    t+=dt
    simulator.step()
```

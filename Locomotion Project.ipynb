```python
import numpy as np
import time

import matplotlib.pyplot as plt

from panda import Simulator, PandaRobot

np.set_printoptions(precision=4, suppress = True)
```

# Helper functions

We provide a set of helper functions (forward kinematics, Jacobians, gravity terms) that can be used to implement the desired controllers. These functions are implemented in the PandaRobot class (in the panda.py file).
Examples are shown below:

```python
# create an instance of the helper class
my_robot = PandaRobot()

# compute the forward kinematics for an arbitrary joint configuration
q = np.random.uniform(-1,1,7)

pose = my_robot.FK(np.array([0.,0,0.0,-np.pi/2.,0.,np.pi/2.,0.0]))
print(f'the pose of the end-effector for joint configuration\n\n {q}\n\n is\n\n {pose}')
```

```python
# we can also get the Jacobian in various frames 
# it is sufficient to pass the name of the frame as an argument as shown below

q = np.random.uniform(-1,1,7)

spatial_jacobian = my_robot.get_jacobian(q, 'S')
body_jacobian = my_robot.get_jacobian(q, 'B')

# we can even get the Jacobian in a frame place at the same position as the end-effector 
# frame but oriented like the spatial frame
oriented_jacobian = my_robot.get_jacobian(q, 'O')

print(f'for joint configuration \n\n {q} \n \n the spatial Jacobian is \n\n {spatial_jacobian}\n\n')
print(f'the body Jacobian is \n\n {body_jacobian}\n\n')
print(f'and the Jacobian in a frame placed at the same position as the end-effector but oriented like the spatial frame is\n\n {oriented_jacobian}')
```

```python
# finally it is possible to compute the effect of gravity on the joints
# this can for example be used to do gravity compensation when controlling the robot

q = np.random.uniform(-1,1,7)

g = my_robot.g(q)

print(f'for joint configuration \n\n {q} \n \n the gravity force seen on the joints is \n \n {g}')
```

# Simulation Code

The class `Simulator` contains all the code for the simulation.

Importantly, we can control both the joints and the gripper of the robot:
* `simulator.send_joint_torque(joint_torques)` sends a vector of dimension 7 setting the torques of the joints
* `simulator.gripper_move(gripper_position)` sends a vector of dimension 2 setting the position of the fingers of the gripper. The fingers can have positions from 0 to 0.04cm (you can fully close the gripper by setting `simulator.gripper_move([0,0])` and fully open it by setting `simulator.gripper_move([0.04,0.04)`)

```python
# this function allows for smooth tracking between the start and goal
# input can be in joint or end effector space and output will match
def compute_trajectory(position_init, position_goal, t_init, t_goal, t):

    T = t_goal-t_init

    desired_position = position_init + ((10/T**3)*(t-t_init)**3 - (15/T**4)*(t-t_init)**4 + (6/T**5)*(t-t_init)**5)*(position_goal-position_init)

    desired_velocity = ((30/T**3)*(t-t_init)**2-(60/T**4)*(t-t_init)**3+(30/T**5)*(t-t_init)**4)*(position_goal-position_init)

    # we return the answer
    return desired_position, desired_velocity
```

```python
# here we create a simulation and reset the state of the robot
simulator = Simulator()
simulator.reset_state([0.,0,0.0,-np.pi/2.,0.,np.pi/2.,0.0])

# we create a robot object so we can use its helper functions
my_robot = PandaRobot()

# duration of the simulation
run_time = 25.

# simulation time step
dt = 0.005
num_steps = int(run_time/dt)

t = 0.

# we store information
ndofs = 7 # number of degrees of freedom (excluding the gripper)
measured_positions = np.zeros([num_steps,ndofs])
measured_velocities = np.zeros([num_steps,ndofs])
desired_torques = np.zeros([num_steps,ndofs])
desired_positions = np.zeros([num_steps,ndofs])
desired_velocities = np.zeros([num_steps,ndofs])
measured_EF_position = np.zeros([num_steps, 3])
measured_EF_velocity = np.zeros([num_steps, 3])
desired_EF_position = np.zeros([num_steps, 3])
desired_EF_velocity = np.zeros([num_steps, 3])
time = np.zeros([num_steps])

# PD gains, tuned to the robot
P = np.array([20.,20.,20.])
D = np.array([60.,60.,60.,60.,60.,60.,60.])

# goal positions and tracking setup; coordinates had to be adjsted to get useful trajectories
bowl = np.array([-0.3,0.55,0.6])
above_block1 = np.array([0.35,0.58,0.4])
block1 = np.array([0.35,0.58,0.3])
above_block2 = np.array([0.15,0.67,0.4])
block2 = np.array([0.15,0.67,0.3])
start_pos = my_robot.FK(simulator.get_state()[0])[:3, 3]
goals = [above_block1, block1, above_block1, bowl, above_block2, block2, above_block2, bowl, start_pos]
ind = 0

### this is the main control loop
for i in range(num_steps):
    # get the current time and save it
    time[i] = dt * i
    
    # we get the position and velocities of the joints
    q, dq = simulator.get_state()
    measured_positions[i,:] = q
    measured_velocities[i,:] = dq
    
    ## TODO: IMPLEMENT CONTROLLER HERE
    # get EF pose through forward kinematics and calculate the jacobian
    pose = my_robot.FK(q)
    jacobian = my_robot.get_jacobian(q, "O") # jacobian of measured joint positions
    
    # save measured EF pos/vel for later plotting
    measured_EF_position[i,:] = pose[:3,3]
    measured_EF_velocity[i,:] = jacobian[3:,:] @ dq
    
    # this controls what the current goal position is
    if time[i] == 0:
        goal = goals[ind]
        start = pose[:3,3]
        time_to_goal = 1.5 * ((goal[0] - start[0])**2 + (goal[1] - start[1])**2 + (goal[2]-start_pos[2])**2)**(0.5) * 5
        
    if t >= time_to_goal:
        start = goals[ind]
        ind += 1
        t = 0
        goal = goals[ind]
        time_to_goal = 1.5 * ((goal[0] - start[0])**2 + (goal[1] - start[1])**2 + (goal[2]-start_pos[2])**2)**(0.5) * 5
    
    # this determines when the gripper opens and closes
    if goal is block1 or goal is block2 or start is bowl:
        gripper_position = [0.04, 0.04]
    else:
        gripper_position = [0.0, 0.0]
    
    # uses the compute_trajectory function to generate smooth paths from start position to each goal
    p_ref, dp_ref = compute_trajectory(start, goal, 0, time_to_goal, t)
    
    # calculations to find the torques to send to joints
    v_o = dp_ref + P * (p_ref - pose[0:3,3])
    jac_inv = np.linalg.pinv(jacobian[3:]) # for desired joint velocity
    dq_des = jac_inv @ v_o
    d_error = dq_des - dq # the velocity error for all the joints
    
    # save the desired EF positions/velocities for later plotting 
    desired_EF_position[i,:] = p_ref
    desired_EF_velocity[i,:] = dp_ref
    desired_velocities[i,:] = dq_des
    
    # we compute the desired torques as a PD controller
    joint_torques = D * d_error + my_robot.g(q)
    desired_torques[i,:] = joint_torques
    
    # we send them to the robot and do one simulation step
    t += dt
    simulator.send_joint_torque(joint_torques)
    simulator.gripper_move(gripper_position)
    simulator.step()
```

```python
def plot_EF_trajectory(measured_EF_position, goals):
    """
    plots the position of the finger in 2D and the position of the spatial frame {s}
    we assume that the time varying x variable is in x_pos and that the y variable is in y_pos
    """
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111)
    plt.plot(measured_EF_position[:,0], measured_EF_position[:,2], 'b')
    plt.xlabel('end effector x position [m]')
    plt.ylabel('end effector z position [m]')
    plt.xlim([-.5, .75])
    plt.ylim([-.2, 1])
    plt.plot([0],[0],'o',markersize=15,color='r')
    
    for goal in goals:
        plt.plot([goal[0]],[goal[2]],'o',markersize=15,color='g')
        if goal is above_block1:
            ax.annotate('above \nblock 1', xy=(goal[0], goal[2]), xytext=(10,0), textcoords='offset points')
        if goal is above_block2:
            ax.annotate('above \nblock 2', xy=(goal[0], goal[2]), xytext=(10,0), textcoords='offset points')
        elif goal is block1:
            ax.annotate('block 1', xy=(goal[0], goal[2]), xytext=(10,0), textcoords='offset points')
        elif goal is block2:
            ax.annotate('block 2', xy=(goal[0], goal[2]), xytext=(10,0), textcoords='offset points')
        elif goal is bowl:
            ax.annotate('above \nbowl', xy=(goal[0], goal[2]), xytext=(10,0), textcoords='offset points')
    ax.annotate('Spatial frame {s}', xy=(-0.03,-0.005), xytext=(30,0), textcoords='offset points')

plot_EF_trajectory(measured_EF_position, [above_block1, block1, above_block2, block2, bowl])
```

```python
def plot_EF_posvel(measured_EF_position, desired_EF_position, measured_EF_velocity, desired_EF_velocity):
    
    fig = plt.figure(figsize = (10, 6))
    
    plt.subplot(3,2,1)
    plt.plot(time, measured_EF_position[:,0], 'b-', time, desired_EF_position[:,0], '--k')
    plt.ylabel(r'$x$')
    plt.subplot(3,2,3)
    plt.plot(time, measured_EF_position[:,1], 'b-', time, desired_EF_position[:,1], '--k')
    plt.ylabel(r'$y$')
    plt.subplot(3,2,5)
    plt.plot(time, measured_EF_position[:,2], 'b-', time, desired_EF_position[:,2], '--k')
    plt.ylabel(r'$z$')
    plt.xlabel('Time [s]')
    
#     fig = plt.figure()
    
    plt.subplot(3,2,2)
    plt.plot(time,measured_EF_velocity[:,0], 'b-', time, desired_EF_velocity[:,0], '--k')
    plt.ylabel(r'$\dot{x}$')
    plt.subplot(3,2,4)
    plt.plot(time, measured_EF_velocity[:,1], 'b-', time, desired_EF_velocity[:,1], '--k')
    plt.ylabel(r'$\dot{y}$')
    plt.subplot(3,2,6)
    plt.plot(time, measured_EF_velocity[:,2], 'b-', time, desired_EF_velocity[:,2], '--k')
    plt.ylabel(r'$\dot{z}$')
    plt.xlabel('Time [s]')
    
plot_EF_posvel(measured_EF_position, desired_EF_position, measured_EF_velocity, desired_EF_velocity)
```
